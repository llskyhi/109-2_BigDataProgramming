```
# 門
table(dataset$A_Phylum)
# 綱
table(dataset$A_Class)
# 目
table(dataset$A_Order)
# 科
table(dataset$A_Family)
# 保育等級
table(dataset$A_Conservation)
```

```
# grep()回傳符合的索引陣列; grepl回傳T或F的陣列, 會比較大
grep("虎", dataset$A_Name_Ch)
grepl("虎", dataset$A_Name_Ch)
```

```
# dataset[列陣列, ]
dataset[grep("虎",dataset$A_Name_Ch), ]
dataset[grepl("虎",dataset$A_Name_Ch), ]
```

```
library(jsonlite)
```

```
url <- "https://data.taipei/opendata/datalist/apiAccess?scope=resourceAquire&rid=f4a75ba9-7721-4363-884d-c3820b0b917c"
petData <- jsonlite::fromJSON(url)
petDataFrame <- petData$result$results
```

```
class(petData)
class(petDataFrame)
dim(petDataFrame)
colnames(petDataFrame)
```

```
print("前3筆")
head(petDataFrame, 3)
print("後3筆")
tail(petDataFrame, 3)
```
