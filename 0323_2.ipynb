```
install.packages("ggplot2")
install.packages("SportsAnalytics")
```

```
library(ggplot2)
library(SportsAnalytics)
```

```
dataNBA <- fetch_NBAPlayerStatistics("19-20")
str(dataNBA)
```

```
dim(dataNBA)
head(dataNBA, 3)
```

```
#qplot(
#  dataNBA$FieldGoalsAttempted,
#  dataNBA$TotalPoints
#  )
qplot(
  FieldGoalsAttempted,
  TotalPoints,
  data=dataNBA,
  main="title"
  )
```

```
qplot(
  data=dataNBA,
  FieldGoalsAttempted,
  TotalPoints,
  color=Position#用打的位置區分點的顏色
  )
```

```
#漸進線?
qplot(
  data=dataNBA,
  ThreesAttempted,
  TotalPoints,
  geom = c("point", "smooth")
  )
```

```
#單變數: 直方圖或長條圖呈現
#fill:填滿長條的顏色 color:長條的外框顏色

#連續資料: 直方圖
qplot(
  data=dataNBA,
  FieldGoalsAttempted,
  fill=Position,#用位置區分bar的顏色
  binwidth=100, #連續bar間距
  color=Team
  )
#類別資料: 長條圖
qplot(
  data=dataNBA,
  Team,
  fill=Position,#用位置區分bar的顏色
  )
```

```
#子圖 facets參數 = 資料~. 或 .~資料
qplot(
  data=dataNBA,
  FieldGoalsAttempted,
  TotalPoints,
  facets = Position~. ,
  color=Team
  )
qplot(
  data=dataNBA,
  FieldGoalsAttempted,
  TotalPoints,
  facets = .~Position ,
  color=Team
  )
```
